```python
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn import datasets, linear_model, metrics 
import pandas as pd
import tensorflow as tf
```

```python
raw_df = pd.read_csv('./data/data.csv')
raw_df
```

```python
#input
#front_df = raw_df[raw_df['type'] == 'front']
front_df = raw_df[raw_df['type'] == 'beside']
x = front_df.iloc[:, 4:].values
x = np.array(x, dtype=float)
#target
y = np.array(front_df.iloc[:, 2].values, dtype=float)
x, y
```

```python
front_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=999999)
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(1, input_dim=5, activation='linear'))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=30000, batch_size=32)
```

```python
model.save('linear_regression_model_beside_3.h5')
```

```python
weights, biases = model.layers[0].get_weights()
print("Weights (Coefficients):", weights)
print("Biases:", biases)
```

```python
loaded_model = tf.keras.models.load_model('linear_regression_model_beside_3.h5')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = loaded_model.predict(x_test)
# TÃ­nh RÂ²
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```

```python
y_pred = loaded_model.predict(x_test)
for id in range(len(y_pred)):
    print(y_pred[id], y_test[id])
```

```python
reg = linear_model.LinearRegression(positive=True) 
reg.fit(x_train, y_train)
```

```python
training_epoch = 10000
learning_rate = 1 / training_epoch
```

```python
reg.coef_
```

```python
reg.score(x_test, y_test)
```

```python
# plot for residual error 
  
# setting plot style 
plt.style.use('fivethirtyeight') 
  
# plotting residual errors in training data 
plt.scatter(reg.predict(x_train), 
            reg.predict(x_train) - y_train, 
            color="green", s=10, 
            label='Train data') 
  
# plotting residual errors in test data 
plt.scatter(reg.predict(x_test), 
            reg.predict(x_test) - y_test, 
            color="blue", s=10, 
            label='Test data') 
  
# plotting line for zero residual error 
plt.hlines(y=0, xmin=0, xmax=50, linewidth=2) 
  
# plotting legend 
plt.legend(loc='upper right') 
  
# plot title 
plt.title("Residual errors") 
  
# method call for showing the plot 
plt.show() 
```
